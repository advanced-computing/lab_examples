```python
# and we will store the data in bigquery
import pandas_gbq
import pydata_google_auth

import yfinance as yf # for downloading stock data
```

### Authentication

```python
# BigQuery authentication
SCOPES = [
    'https://www.googleapis.com/auth/cloud-platform',
    'https://www.googleapis.com/auth/drive',
]

credentials = pydata_google_auth.get_user_credentials(
    SCOPES,
    # Note, this doesn't work if you're running from a notebook on a
    # remote sever, such as over SSH or with Google Colab. In those cases,
    # install the gcloud command line interface and authenticate with the
    # `gcloud auth application-default login` command and the `--no-browser`
    # option.
    auth_local_webserver=True,
)
```

### Getting daily data from last month

```python
dat = yf.Ticker("MSFT")
msft_df = dat.history(period='1mo')
# drop last 2 rows, so we can update later
msft_df = msft_df[:-2]
msft_df
```

## Creating BigQuery Table

```python
# to BigQuery
pandas_gbq.to_gbq(
    msft_df,
    destination_table='stock_data.msft',
    project_id='sipa-adv-c-roberto',
    if_exists='replace',
    credentials=credentials,
)
```
