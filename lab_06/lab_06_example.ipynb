```python
import numpy as np
import pandas as pd

def simulate_data(n):
    x = np.random.uniform(0, 1, n)
    y = 2 + 3 * x + np.random.normal(0, 1, n)
    return x, y

from matplotlib import pyplot as plt

def plot_data(x, y):
    width = 100
    height = 100
    plt.scatter(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


plot_data(*simulate_data(100))
```
